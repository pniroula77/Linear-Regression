### Function to calculate cost and Gradient

```python
def gradient_fwd(X,Y,W,b):
    '''
    parameters:
    X:Matrix consisting of features
    Y:Matrix containing actual price 
    W:Weights(that needs to be adjusted)
    b:bias term
    
    returns:
    error: error term(Sum of Square of error is used here.)
    dW:gradient descent for weight matrix
    db:gradient descent for bias term
    
    '''
    y_pred = b + np.dot(X,W.T)
    y_actual = Y.values.reshape(Y.shape[0], 1)
    
    error = np.sum(y_pred-y_actual)**2
    
    dW = -2* np.dot((y_actual - y_pred).T, X)
    
    db = -2*np.sum(y_actual-y_pred)
    
    return error,dW,db
    
# a,b = gradient_fwd(X_train,y_train,np.random.random((1,X_train.shape[1])),0)
# a.shape
```

### Function to adjust weight

```python
def adjust_wt(X,Y,W,b,n_iter,learning_rate):
    '''
    parameters:
    X:Matrix of Features
    Y:Matrix containing actual price
    W:Weights(to be adjusted)
    b:bias(to be adjusted)
    n_iter: no. of iterations
    
    returns:
    error: error term(Sum of Square of error is used here.)
    W:Adjusted Weight Matrix
    b:Adjusted bias Term
    
    '''
    errors = []

    for i in range(n_iter):
        # Cost and gradient calculation call function propagate
        err,dW,db = gradient_fwd(X, Y,W,b)

        

        # update parameters
        W = W - learning_rate * dW
        b = b - learning_rate * db

        errors.append(err)


    return errors, W,b
```

### Function that Predicts the value

```python
def calculate(X,W,b):
    '''
    parameters:
    X: Matrix of Features
    B: Weights
    b:bias term
    
    returns:
    predicted_value: The value computed from the parameters
    '''
    predicted_value = b + np.dot(X,W.T)
    return predicted_value
```

### Actual Model that implements linear regression

```python
def linear_regression(X_train, y_train, X_test, y_test, iter=2000, learning_rate=0.5):
    '''
    parameters:
    X_train: Matrix of features that is used for training(80% of total data of dataset)
    y_train: Matrix that contains the actual price used for model training(80% of total data of dataset)
    X_test: Matrix of features that is used for testing model(20% of total data of dataset)
    y_test: Matrix that contains the actual price used for model testing(20% of total data of dataset)
    
    returns:
    result: A dictionary containing all the information of model
    '''
     # randomly initialize the parameters
    W = np.random.random((1, X_train.shape[1]))
    b = 0
     
    # Gradient descent
    error,weight,bias = adjust_wt(X_train, y_train,W,b, iter, learning_rate)
    
    # Calculate the predicted value
    Y_prediction_test = calculate(X_test,W,b)
    Y_prediction_train = calculate(X_train,W,b)
    
    #actual values
    Y_train = y_train.values.reshape(y_train.shape[0], 1)
    Y_test = y_test.values.reshape(y_test.shape[0], 1)
    
    # Print train/test Errors
    print("train accuracy: {} %".format(100 - np.mean(np.abs(Y_prediction_train - Y_train)) * 100))
    print("test accuracy: {} %".format(100 - np.mean(np.abs(Y_prediction_test - Y_test)) * 100))

    result = {"errors": error,
         "Y_prediction_test": Y_prediction_test,
         "Y_prediction_train": Y_prediction_train,
         "W": W,
         "b": b,
         "learning_rate": learning_rate,
         "num_iterations": iter}

    return result
```
